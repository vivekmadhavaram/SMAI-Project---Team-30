```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files=["subject102.DAT","subject103.DAT","subject104.DAT","subject105.DAT","subject106.DAT","subject107.DAT","subject108.DAT","subject109.DAT"]
data=pd.read_csv(r"C:\Users\vivek\Downloads\SMAI\Project\PAMAP2_Dataset\Protocol\subject101.DAT",sep=" ",header=None)
data=data.interpolate(method ='linear', limit_direction ='forward').copy()
for i in range(len(files)):
    path="C://Users//vivek//Downloads//SMAI//Project//PAMAP2_Dataset//Protocol//"+files[i]
    temp=pd.read_csv(path,sep=" ",header=None)
    temp=temp.interpolate(method ='linear').copy()
    data=data.append(temp, ignore_index=True)
  
```

```python
data.rename(columns = {1:'activity'}, inplace = True)
data
```

```python
data=data.dropna() 
data.shape

activity_id = {1:'lying', 2:'sitting', 3:'standing',
              4:'walking', 5:'running', 6:'cycling', 7:'Nordic walking', 
              12:'ascending stairs', 13:'descending stairs', 16:'vacuum cleaning',
              17:'ironing', 24:'rope jumping'}
data=data[data['activity'].isin(activity_id.keys())]
data=data[data.columns[:-34]]
data
```

```python
from sklearn.model_selection import train_test_split
y=data[data.columns[1]]
X=data[data.columns[2:]]
for i in X.columns:
    X[i] = (X[i] - X[i].mean()) / X[i].std()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
X_test
```

```python
tx1=X_train[:int(len(X_train)/5)].copy()
tx2=X_train[int(len(X_train)/5):int(2*len(X_train)/5)].copy()
tx3=X_train[int(2*len(X_train)/5):int(3*len(X_train)/5)].copy()
tx4=X_train[int(3*len(X_train)/5):int(4*len(X_train)/5)].copy()
tx5=X_train[int(4*len(X_train)/5):].copy()
ty1=y_train[:int(len(y_train)/5)].copy()
ty2=y_train[int(len(y_train)/5):int(2*len(y_train)/5)].copy()
ty3=y_train[int(2*len(y_train)/5):int(3*len(y_train)/5)].copy()
ty4=y_train[int(3*len(y_train)/5):int(4*len(y_train)/5)].copy()
ty5=y_train[int(4*len(y_train)/5):].copy()
training_x=[tx1,tx2,tx3,tx4,tx5]
training_y=[ty1,ty2,ty3,ty4,ty5]
```

```python
def data(i):
    temp_x=pd.DataFrame()
    temp_y=pd.DataFrame()
    val_x=pd.DataFrame()
    val_y=pd.DataFrame()
    for j in range(5):
        if (i!=j):
            temp_x=pd.concat([temp_x,training_x[j]])
            temp_y=pd.concat([temp_y,training_y[j]])
        else:
            val_x=pd.concat([val_x,training_x[j]])
            val_y=pd.concat([val_y,training_y[j]])
    return temp_x,temp_y,val_x,val_y
```

```python
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import time
scores=pd.DataFrame(columns=["Fold.no","Training Score","Validation Score","Test Score"])
classifiers=[]
#pca = PCA(n_components=17)
for i in range(5):
    data_x,data_y,v_x,v_y=data(i)
    
    start_time = time.time()
    #x1=pca.fit_transform(data_x)
    #x2=pca.fit_transform(v_x)
    #x3=pca.fit_transform(X_test)
    classifier = DecisionTreeClassifier(max_depth=15)
    classifier.fit(data_x, data_y)
    vy_pred = classifier.predict(v_x)
    sc2=accuracy_score(v_y, vy_pred)
    try_pred = classifier.predict(data_x)
    sc1=accuracy_score(data_y, try_pred)
    tey_pred = classifier.predict(X_test)
    sc3=accuracy_score(y_test, tey_pred)
    classifiers.append(classifier)
    print(time.time()-start_time)
    scores.loc[len(scores)]=[i,sc1,sc2,sc3]
    print(i)
print("Best Test Accuracy is :", max(scores['Test Score']))
scores
```

```python
from sklearn.metrics import confusion_matrix
cls=classifiers[0]
y_t=list(y_test)
tey_pred = cls.predict(X_test)
cm2 = confusion_matrix(y_t, list(tey_pred))
cm_df2 = pd.DataFrame(cm2,index = activity_id.values(),columns = activity_id.values())
print("Confusion Matrix")
cm_df2
```
